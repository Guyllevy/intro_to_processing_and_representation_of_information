```python
# get image
# turn image into array
from PIL import Image
import numpy as np
 
img = Image.open('image.png')
np_img = np.asarray(img)

print(np_img.shape)
```

```python
# turn image array to gray scale
np_img_gs = np.mean(np_img, axis = 2, dtype = np.int64)
print(np_img_gs.shape) 
print(np_img_gs)
print("max value:",np_img_gs.max())
print("min value:",np_img_gs.min())
```

```python
import matplotlib.pyplot as plt
plt.imshow(np_img_gs)
```

```python
# (1)
# estimate pdf of gray levels with histogram
hist = {value:0 for value in range(256)}
for value in np_img_gs.flatten():
    hist[value] += 1
plt.title("histogram (unnormalized pdf) of gray levels", fontdict=None, loc='center')
plt.bar(list(hist.keys()), hist.values())
plt.show()
```

```python
# (2)
import matplotlib.pyplot as plt

img = np_img_gs
rows, cols = 3, 3
MSEs = []
Decisions = []
Representations = []

for b in range(1, 9):
    plt.subplot(rows, cols, b)
    b_img = img
    
    Ds, Rs = [], []
    for i in range(2**b):
        d_begin = int(255 * (i / 2**b))
        d_end   = int(255 * ((i + 1) / 2**b))
        r = int((d_end + d_begin)/2)
        b_img = np.where((b_img >= d_begin) & (b_img <= d_end),r , b_img)
        Ds.append(d_end)
        Rs.append(r)
        
    mse = ((b_img - img)**2).mean()
    MSEs.append(mse)
    Decisions.append(Ds)
    Representations.append(Rs)
    
    plt.title(f"{b} bits quantization, mse:{mse:.2e}", fontdict=None, loc='center', fontsize = 8)
    plt.imshow(b_img)

plt.tight_layout()
plt.show()
```

```python
plt.title("mse as function of b", fontdict=None, loc='center')
plt.plot(MSEs)
plt.show() # (a) ploting mse as a function of b
```

```python
for b in range(1, 9):
    plt.subplot(rows, cols, b)
    b_img = img

    plt.title(f"representation and decision levels b = {b}", fontdict=None, loc='center', fontsize = 7)
    plt.vlines(x = Decisions[b-1], ymin = 0, ymax = 255)
    plt.hlines(y= Representations[b-1], xmin = [0] + Decisions[b-1][:-1], xmax= Decisions[b-1], colors='purple')
    plt.xlim(0,255)
    plt.ylim(0,255)
    
plt.tight_layout()
plt.show() # (b) plotting represantation and decisions levels for each b 
```

```python
from math import inf

def calc_mse(img,Ds,Rs,bits):
    b_img = img
    for i in range(2**bits):
        d_begin = Ds[i]
        d_end   = Ds[i+1]
        r       = Rs[i]
        b_img   = np.where((b_img >= d_begin) & (b_img <= d_end),r , b_img)
    return ((b_img - img)**2).mean()

def Max_Lloyd(img, pdf, initial_decisions, eps, bits):
    
    Ds = np.array([0] + initial_decisions + [255])     #   0  [d0  d1  d2  d3  d4  d5  d6]  255
    Rs = np.zeros(len(initial_decisions) + 1)          #    [r0  r1  r2  r3  r4  r5  r6  d7]
    for i in range(len(Rs)):
        Rs[i] = (Ds[i] + Ds[i+1])/2

    mse_prev = inf
    mse = calc_mse(img, Ds, Rs, bits)
    MSEs = [mse]
    
    while  abs(mse_prev - mse) > eps:
        
        for i in range(len(Ds)-1):
            Xs = np.array(list(range(Ds[i],Ds[i+1])))
            probs = np.array(pdf[Ds[i]:Ds[i+1]])
            Rs[i] = (probs * Xs).sum()
            Rs[i] /= probs.sum()
            
        for j in range(1,len(Ds)-1):
            Ds[j] = (Rs[j-1] + Rs[j]) / 2
            
        mse_prev = mse
        mse = calc_mse(img, Ds, Rs, bits)
        MSEs.append(mse)
        
    Decisions = [int(f) for f in Ds.tolist()[1:-1]] # not including 0, 255
    Representations = [int(f) for f in Rs.tolist()]
    
    print(MSEs)
    return (Decisions, Representations)
```

```python
img = np_img_gs
pdf = list(hist.values())
bits = 4
initial_decisions = []
for i in range(2**bits -1):
    d_end   = int(255 * ((i + 1) / 2**bits))
    initial_decisions.append(d_end)
eps = 0.01

Max_Lloyd(img, pdf, initial_decisions, eps, bits)
```

```python
def show_quantized_image(img,Ds,Rs):
    Ds = [0] + Ds + [255]
    b_img = img
    for i in range(len(Rs)):
        b_img = np.where((b_img >= Ds[i]) & (b_img <= Ds[i+1]), Rs[i] , b_img)
    plt.imshow(b_img)
    plt.show()
```

```python
show_quantized_image(img, *Max_Lloyd(img, pdf, initial_decisions, eps, bits))
```
